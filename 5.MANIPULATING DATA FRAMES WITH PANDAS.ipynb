INDEXING DATA FRAMES

```python
import pandas as pd

# read data
data = pd.read_csv('pokemon.csv')
data= data.set_index("#") # index sütunu istediğimiz bir sütun yapabiliriz 
data.head()
```

```python
# indexing using square brackets
data["HP"][1]
```

```python
# using column attribute and row label
print(data.HP[1])

# aşağıdaki benzer fonksiyonlarla da istenilen değeri seçebiliriz 
print(type(data.loc[1,"HP"]))
print(data.HP.at[1])
print(data.HP.iloc[0])
print(data.iloc[0,3])
```

```python
from IPython.display import display

# using loc accessor
print(data.loc[1,["HP"]]) # bu örnek için [] kullandık ve tipi series oldu eğer kullanmasaydık tipi integer olacaktı  

display(data.loc[(1,2,3,15),["HP"]]) #bu örnek için [] kullandık ve tipi DataFrame oldu eğer kullanmasaydık tipi series olacaktı
```

```python
# Selecting only some columns
display(data[["HP","Attack"]])
```

SLICING DATA FRAME

```python
# Difference between selecting columns: series and dataframes
print(type(data["HP"]))     # series
print(type(data[["HP"]]))   # data frames
```

```python
# Slicing and indexing series
display(data.loc[1:10:1,"HP":"Defense"])  # 10 and "Defense" are inclusive


c_names={}

# ortalaması alınabilecek sutünların isim ve ortalamalarını dictionary olarak alıyoruz 
for colomn in data.columns:    
    mean=data[[colomn]].mean(numeric_only=True).to_list()
    if ( mean ):
        c_names[colomn]=mean

# bu yaptığım dictionary kullanarak bir dataframe oluşturabiliriz 
mean_table=pd.DataFrame(c_names)

display(mean_table)

# ortalaması en büyük 3 sütünün ilk 20 elemanını ayırıyoruz 
display(data.loc[1:20:1,sorted(c_names,key=c_names.get,reverse=True)[0:3]])

```

```python
# From something to end
display(data.loc[1:10,"Speed":])
```

FILTERING DATA FRAMES

```python
# Creating boolean series
boolean = data.HP > 200
data[boolean]
```

```python
# Combining filters
first_filter = data.HP > 150
second_filter = data.Speed > 35
third_filter = data.Generation >1
data[first_filter & second_filter & third_filter]
```

```python
# Filtering column based others
data.HP[data.Speed<15]
```

TRANSFORMING DATA

```python
# Plain python functions
def div(n):
    return n/2

display(data.HP.apply(div)) # hp sutünundaki her eleman için verilen fonksiyonu uygular 
display(data.HP)
```

```python
# Or we can use lambda function
data.HP.apply(lambda n : n/2) # üst kod ile aynı şekilde verilere istenilen fonksiyou uygular 
```

```python
# Defining column using other columns
data["total_power"] = data.Attack + data.Defense
display(data.head())

data["combat_power"] = data.HP * data.Speed
display(data.head())
```

HIERARCHICAL INDEXING

```python
# lets read data frame one more time to start from beginning
data = pd.read_csv('pokemon.csv')
data.head()
# As you can see there is index. However we want to set one or more column to be index
```

```python
# Setting index : type 1 is outer type 2 is inner index
data1 = data.set_index(["Type 1","Type 2"]) 
display(data1.head())
# data1.loc["Fire","Flying"] # howw to use indexes

data.reset_index(drop=True)
data.head()
```

PIVOTING DATA FRAMES

```python
dic = {"treatment":["A","A","B","B"],"gender":["F","M","F","M"],"response":[10,45,5,9],"age":[15,4,72,65]}
df = pd.DataFrame(dic)
df
```

```python
# pivoting
display(df.pivot(index="treatment",columns = "gender",values="response"))

new_data=data.groupby("Type 1").mean().drop(columns='#')
new_data.reset_index(inplace=True)

display(new_data.pivot(index="Type 1",columns='Speed',values='HP')) # not logical but we can do something like this 
```

STACKING and UNSTACKING DATAFRAME

```python
df1 = df.set_index(["treatment","gender"])
display(df1)
# lets unstack it
data=data.set_index(["Type 1","Type 2"])
display(data.head())
```

```python
# level determines indexes
display(df1.unstack(level=0))
```

```python
display(df1.unstack(level=1))
```

```python
# change inner and outer level index position
df2 = df1.swaplevel(0,1)
display(df2)

display(data.swaplevel(0,1)) # Type 2 ve type 1 in yerlerini değiştirdi

```

CATEGORICALS AND GROUPBY

```python
# We will use df
df
```

```python
# according to treatment take means of other features
display(df.groupby("treatment").mean() )  # mean is aggregation / reduction method
# there are other methods like sum, std,max or min

display(data.groupby("Type 2").sum())
```

```python
# we can only choose one of the feature
display(df.groupby("treatment").age.max() )

display(data.groupby("Type 2").Attack.max())
```

```python
# Or we can choose multiple features
display(df.groupby("treatment")[["age","response"]].min() )

display(data.groupby("Type 2")[["Attack","Defense"]].max())
```
