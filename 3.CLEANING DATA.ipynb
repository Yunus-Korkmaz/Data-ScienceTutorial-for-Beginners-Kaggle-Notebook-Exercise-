DATA TYPES

```python
# 5 temel veri tipimiz vardır : object(string), integer, boolen, float and categorical
# verileri categorical tipe çevirmek verinin boyutunu küçültür
data.dtypes
```

```python
# lets convert object(str) to categorical and int to float.
original_data=data.copy()
data['Type 1'] = data['Type 1'].astype('category')
data['Name'] = data['Name'].astype('category')
data['Speed'] = data['Speed'].astype('float')
```

```python
import sys
# As you can see Type 1 is converted from object to categorical
# And Speed is converted from int to float

dtypes_of_datas=pd.concat([data.dtypes,original_data.dtypes],axis=1)
display(dtypes_of_datas)
a=5
print('original data size = ',sys.getsizeof(original_data),'\nconverted data size] = ',sys.getsizeof(data))
# arada boyut farkı var 
```

```python
# Lets look at does pokemon data have nan value
# As you can see there are 800 entries. However Type 2 has 414 non-null object so it has 386 null object.
data.info()
```

```python
# Lets chech Type 2
data["Type 2"].value_counts(dropna =False)
# As you can see, there are 386 NAN value
```

```python
# Lets drop nan values
data1=data   # also we will use data to fill missing value so I assign it to data1 variable

# data1["Type 2"].dropna(inplace = True) çalışmadı alttaki benzer kod çalıştı 
data1.dropna(subset=['Type 2'],inplace = True)  # inplace = True means we do not assign it to new variable. Changes automatically assigned to data
# So does it work ?


data1["Type 2"].value_counts(dropna =False)
```
